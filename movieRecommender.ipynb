# Importing the Data and combining the two datasets

```python
import numpy as np
import pandas as pd
import nltk
```

```python
movies=pd.read_csv('tmdb_5000_movies.csv')
credits=pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies=movies.merge(credits,on='title')
```

# Sifting out important columns

```python
#genres,id, keywords,title, overview, cast, crew

movies=movies[['movie_id','title','genres','keywords','overview','cast','crew']]
```

```python
movies.head()
```

# Preprocessing Data

```python
movies.isnull().sum()
```

```python
movies=movies.dropna()
movies
```

```python
movies.isnull().sum()
```

```python
movies.duplicated().sum()
```

```python
movies['genres'].head(1).values
```

```python
import ast

def convert(listOfDics):
    genresList=[]
    for dic in ast.literal_eval(listOfDics):
        genresList.append(dic["name"])
    return genresList
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head(4)
```

```python
def convertCast(listOfDics):
    castList=[]
    i=0
    for dic in ast.literal_eval(listOfDics):
        if i>2:
            break
        castList.append(dic['name'])
        i+=1
    return castList
    
```

```python
movies['cast']=movies['cast'].apply(convertCast)
```

```python
movies.head()
```

```python
def findDir(listOfDics):
    L=[]
    for dic in ast.literal_eval(listOfDics):
        if dic['job']=='Director':
            L.append(dic['name'])
            break
    return L
```

```python
movies['crew']=movies['crew'].apply(findDir)
```

```python
movies.head()
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

## Removing Whitespaces

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies.head()
```

```python
movies['tags']=movies['overview']+movies['keywords']+movies['genres']+movies['cast']+movies['crew']
movies=movies.drop(['overview','keywords','genres','cast','crew'],axis=1)
```

```python
movies.head()
```

```python
movies.head(1).values
```

```python
new_df=movies
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head(2)
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df['tags'][0]
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

# Text Vectorization using Bag of Words

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
 vectors
```

```python
vectors[0]
```

```python
print(cv.get_feature_names_out())
```

```python
vectors[0]
```

```python
#since there are 5000 dimensions to our vector, we wont use euclidian distance, as that is not a good measure!
#instead we will use cosine distance which is basically the angle between two vectors
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vectors)
```

```python
def recommend(movie):
    movie_index=new_df[new_df['title']==movie].index[0]
    distances=similarity[movie_index]
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
